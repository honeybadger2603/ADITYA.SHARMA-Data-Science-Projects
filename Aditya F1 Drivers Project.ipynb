# **PART II) HYPOTHESIS TESTING

```python
import scipy.stats as stats
```

```python
#Hypothesis - Average podium rate of F1 drivers is 0.05 or 5%

H0: meu  = 0.05 or 5%
HA: meu != 0.05 or 5%
```

```python
df['Podium_Rate'].describe()
```

```python
def z_score(pp, pop_mean, pop_sd):
    z = (pp.mean() - pop_mean) / (pop_sd / (len(pp) ** 0.5))
    print (z, stats.norm.cdf(z))
```

```python
z_score(df['Podium_Rate'], 0.05, 0.1)
```

```python
stats.ttest_1samp(df.Podium_Rate, 0.05)
```

**Since the p-value is < 0.05 (significance level), reject null hypothesis and accept the alternate hypothesis - 
average podium rate of f1 drivers is not 0.05 or 5%

```python
#Hypothesis -  The average number of race starts of drivers in F1 is 27

H0: meu  = 27
HA: meu  < 27
```

```python
df['Race_Starts'].describe()
```

```python
def z_score(pp, pop_mean, pop_sd):
    z = (pp.mean() - pop_mean) / (pop_sd / (len(pp) ** 0.5))
    print (z, stats.norm.cdf(z))
```

```python
z_score(df['Race_Starts'], 27, 51.28)
```

```python
stats.ttest_1samp(df.Race_Starts, 27)
```

**Since the p-value is > 0.05 (level of significance), we fail to reject null hypothesis
and therefore accept that average race starts by all drivers in F1 is 27

```python
#Hypothesis - Average fastest lap of the drivers is equal to the average win rate of the drivers

H0: Avg. Fastest Lap  = Avg. Win Rate
HA: Avg. Fastest Lap != Avg. Win Rate
```

```python
df1 = df.dropna(subset = ["Win_Rate", "Fastest_Laps"])
```

```python
stats.ttest_ind(df1['Win_Rate'], df1['Fastest_Laps'])
```

**2.3081785207145172e-07 = 0.000000230817852071452 and since its is < 0.05, we reject the null hypothesis thus proving that
HA Average of fastest Lap is not equal to Average of win rate. By this we mean that just because a driver gets fastest lap in a race 
it does not improve his race winning rate

```python
#Hypothesis - Avg. Criteria Race_Starts and Avg. Points are equal

H0: Avg. of Race_Starts = Avg. Points scored 
HA: Avg. of Race_Starts != Avg.Points scored 
```

```python
df2 = df.dropna(subset = ["Race_Starts", "Points"])
```

```python
stats.ttest_ind(df2['Race_Starts'], df2['Points'])
```

Since the value of p here is greater than 0.05 (the significance level), we fail to reject the null hypothesis,
and therefore agree that the criteria Avg. Race starts and Avg. Point scored are equal to one another

```python
#Here is a chi-square test to determine the relationship between a drivers nationality and the number of championships he has won
```

#Hypothesis: There's a relationship between a driver's nationality and the number of championships he has won

H0: There's no relationship between the drivers' nationality and number of championships won.
HA: There's a relationship between the drivers' nationality and number of championships won.

```python
from scipy.stats import chi2_contingency
# I have created a cross table for 'Nationality' and 'Championships' columns
cross_table = pd.crosstab(df['Nationality'], df['Championships'])
chi2, p, dof, expected = chi2_contingency(cross_table)
```

```python
print(f"Chi-Square: {chi2}")
print(f"p-value: {p}")
```

Interpretation: In this case, the P-value attained through the chi-squared test is higher than the significance level
of 0.05 or 5%. Hence, we fail to reject the null hypothesis thus confirming that there is no relevant relationship that
exists between the nationality of a racer and the number of championships he has won.
